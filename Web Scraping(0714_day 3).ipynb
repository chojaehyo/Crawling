#### [네이버 로그인 전체 코드]

```python
import time
from selenium import webdriver

browser = webdriver.Chrome()

# 1. 네이버 이동
browser.get("http://naver.com")
```

```python
# 2. 로그인 버튼 클릭
elem = browser.find_element_by_class_name("link_login")
elem.click()

# 3. id, pw 입력
elem = browser.find_element_by_id("id")
elem.send_keys("my_id")
time.sleep(3) # 3초간 정지

elem = browser.find_element_by_id("pw")
elem.send_keys("my_pw")
time.sleep(3)

# 4. 로그인 버튼 클릭
elem = browser.find_element_by_id("log.login")
elem.click()

# 5. 페이지 소스 보기
print(browser.page_source)
time.sleep(3)

# 6. 브라우저 종료
browser.quit()
```

#### [네이버 항공권 검색 전체 코드]

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
browser = webdriver.Chrome()
browser.maximize_window()  # 시작할 때 브라우저 최대화

# 1. 네이버 항공권 예매 사이트
url = "https://flight.naver.com/flights/"
browser.get(url)

# 2. 가는날/오는날 선택
browser.find_element_by_link_text("가는날 선택").click()

browser.find_elements_by_link_text("27")[0].click()
browser.find_elements_by_link_text("28")[0].click()

# 3. 재주도 선택
browser.find_element_by_xpath('//*[@id="recommendationList"]/ul/li[1]/div/span').click()

# 4. 항공권 검색
browser.find_element_by_link_text("항공권 검색").click()

# 5. 검색 기다리기 (xpath에 해당하는 element가 나올 때까지 기다리기)
elem = WebDriverWait(browser, 10).until(EC.presence_of_element_located((By.XPATH, '//*[@id="content"]/div[2]/div/div[4]/ul/li[1]')))

# 6. 결과 출력하기                                                                       
print(elem.text)
                                                                       
# 7. 브라우저 종료
browser.close()                                                                    
```

#### - 문서의 height 가지고 페이지 스크롤하기

```python
browser.execute_script("window.scrollTo(0, document.body.scrollHeight )")
```

```python
prev_height = browser.execute_script("return document.body.scrollHeight")

while True:
    # 스크롤 내리기
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight )")
    
    time.sleep(2)
    # 현재 height 구하기
    curr_height = browser.execute_script("return document.body.scrollHeight")
    
    if curr_height == prev_height:
        break
        
    prev_height = curr_height

print("스크롤 완료")
```

#### - 스크롤된 모든 데이터에 대해

```python
soup = BeautifulSoup(browser.page_source, 'lxml')

movies = soup.find_all('div', {'class': 'Vpfmgd'})
print('전체 영화 수:', len(movies))

for movie in movies:
    if movie.find('span', {'class':'SUZt4c djCuy'}) != None:
        name = movie.find('div', {'class': 'WsMG1c nnK0zc'}).get_text()
        pre_price = movie.find('span', {'class':'SUZt4c djCuy'}).get_text()
        cur_price = movie.find('span', {'class':'VfPpfd ZdBevf i5DZme'}).get_text()
        link = "https://play.google.com" + movie.find('a')['href']
        print('영화제목:' ,name)
        print('할인 전 가격:', pre_price)
        print('할인 후 가격:', cur_price)
        print('링크:', link)
        print("="*100)
```

#### - 딸기 이미지 10개 저장하기

```python
from selenium import webdriver
import time
import urllib.request

browser = webdriver.Chrome()
browser.maximize_window()

url = "https://www.google.co.kr/imghp?hl=ko"
browser.get(url)

elem = browser.find_element_by_name("q")
elem.clear()
elem.send_keys("딸기")
elem.submit()

prev_height = browser.execute_script("return document.body.scrollHeight")

while True:
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight )")
    time.sleep(2)
    curr_height = browser.execute_script("return document.body.scrollHeight")
    if curr_height == prev_height:
        break
    prev_height = curr_height

print("스크롤 완료")

images = browser.find_elements_by_css_selector(".rg_i.Q4LuWd")
count = 1
for image in images:
    image.click()
    time.sleep(2)
    imgUrl = browser.find_element_by_css_selector(".n3VNCb").get_attribute("src")
    
    urllib.request.urlretrieve(imgUrl, str(count) + ".jpg")
    count = count + 1
    if count > 10:
        break
        
time.sleep(2)
browser.close()
```
